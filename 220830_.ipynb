```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
#1, 11, 21, 31...
from tqdm import tqdm

total = []

for n in tqdm(range(1, 3992, 10)):

    url = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%9D%B4%ED%99%94%EC%97%AC%EB%8C%80&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=46&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={}".format(n))
    html = BeautifulSoup(url.text)
    
    #find_all > 리스트로 반환
    news = html.find("table", class_="type")

    for i in news:
        title = i.find("a", class_="news_tit").text
        content = i.find("div", class_="dsc_wrap").text
        total.append([title, content])
    
    time.sleep(1)
```

```python
df = pd.DataFrame(total, columns = ["제목", "내용"])
df
```

```python
t = ''
for i in total:
    t += i[0] + ' ' + i[1] + ' '
len(t)
```

```python
#머신러닝: 병렬계산
import numpy as np

x1 = np.array([2, 4, 6, 8]) #공부 시간
x2 = np.array([1, 2, 1, 3]) #문제집 푼 개수
y = np.array([35, 75, 60, 90]) #성적

a1 = 0
a2 = 0
b = 0
lr = 0.005

for i in range(3000):
    y_hat = a1 * x1 + a2 * x2 + b
    error = y_hat - y
    
    a1_diff = sum(2 * error * x1)
    a2_diff = sum(2 * error * x2)
    b_diff = sum(2 * error * 1)

    a1 = a1 - a1_diff * lr
    a2 = a2 - a2_diff * lr
    b = b - b_diff * lr

    print("{}회 학습 // 기울기: {} // 절편: {}".format(i+1, a1, a2, b) )
```

```python
plt.scatter(x1, y)
plt.plot(x1, a1*x1 + a2*x2 + b, "red")
```
